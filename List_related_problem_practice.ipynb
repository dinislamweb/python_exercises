1. Write a code to create a new list using odd-indexed elements from the first list and even-indexed elements from the second

- Given two lists, l1 and l2, write a program to create a third list l3 by picking an odd-index element from the list l1 and even index elements from the list l2.

- l1 = [3, 6, 9, 12, 15, 18, 21]
- l2 = [4, 8, 12, 16, 20, 24, 28]


Expected Output:

Element at odd-index positions from list one
[6, 12, 18]

Element at even-index positions from list two
[4, 12, 20, 28]

Printing Final third list
[6, 12, 18, 4, 12, 20, 28]

```python
l1 = [3, 6, 9, 12, 15, 18, 21]

l2 = [4, 8, 12, 16, 20, 24, 28]

odd_indexed = l1[1::2]
even_indexed = l2[::2]
l3 = []
l3.extend(odd_indexed)
l3.extend(even_indexed)
print(l3)
```

Write a program to remove the item present at index 4 and add it to the 2nd position and at the end of the list.

Given:

sample_list = [34, 54, 67, 89, 11, 43, 94]

Expected Output:

- List After removing element at index 4  [34, 54, 67, 89, 43, 94]
- List after Adding element at index 2  [34, 54, 11, 67, 89, 43, 94]
- List after Adding element at last  [34, 54, 11, 67, 89, 43, 94, 11]

```python
sample_list = [34, 54, 67, 89, 11, 43, 94]

print("Original list ", sample_list)
element = sample_list.pop(4)
print("List After removing element at index 4 ", sample_list)

sample_list.insert(2, element)
print("List after Adding element at index 2 ", sample_list)

sample_list.append(element)
print("List after Adding element at last ", sample_list)
```

Slice list into 3 equal chunks and reverse each chunk
- Given:

- sample_list = [11, 45, 8, 23, 14, 12, 78, 45, 89]
- Expected Outcome:

- Chunk  1 [11, 45, 8]
- After reversing it  [8, 45, 11]
- Chunk  2 [23, 14, 12]
- After reversing it  [12, 14, 23]
- Chunk  3 [78, 45, 89]
- After reversing it  [89, 45, 78]

```python
sample_list = [11, 45, 8, 23, 14, 12, 78, 45, 89]
length = len(sample_list)
chunk_size = int(length/3)
start = 0
end = chunk_size
for i in range(3):
    indexes = slice(start,end)
    list_chunk = sample_list[indexes]
    print("Chunk",i+1,list_chunk)
    print("After reversing it",list(reversed(list_chunk)))
    start = end
    end+=chunk_size
    
```

4. Write a program to iterate a given list and count the occurrence of each element and create a dictionary to show the count of each element

sample_list = [11, 45, 8, 11, 23, 45, 23, 45, 89]
- Expected Output:

Printing count of each item   {11: 2, 45: 3, 8: 1, 23: 2, 89: 1}

```python
from collections import Counter
sample_list = [11, 45, 8, 11, 23, 45, 23, 45, 89]
char_dict = Counter(sample_list)

print("Printing count of each item:", dict(char_dict))
```

```python
sample_list = [11, 45, 8, 11, 23, 45, 23, 45, 89]
new_dict = dict()
for char in sample_list:
    count = sample_list.count(char)
    new_dict[char] = count
print("Printing count of each item:",new_dict)
```

Calculate and print the sum and average of all numbers in a list.

Given:

1. my_list = [10, 20, 30, 40, 50]
- Expected Output:
- Sum: 150
- Average: 30.0

```python
my_list = [10, 20, 30, 40, 50]
sum = 0
for i in my_list:
    sum+=i
avg = sum/len(my_list)
print(f"Sum: {sum}")
print(f"Average: {avg}")
```

Reverse a list

Given:

1. list1 = [100, 200, 300, 400, 500]
- Expected output:

[500, 400, 300, 200, 100]

```python
list1 = [100, 200, 300, 400, 500]
rev_list = list1[::-1]
print(rev_list)
```

```python
list1 = [100, 200, 300, 400, 500]
list1.reverse()
print(list1)
```
