- Tuple1 = (1, 2, 3)

- Tuple 1: (1, 2, 3, 8)

- (1, 2, 3, 8) is less than (1, 2, 4, 5)

```python
Tuple1 = (1, 2, 3, 8)

Tuple2 =  (1, 2, 4, 5)

print(f"Tuple 1: {Tuple1}")
print(f"Tuple 2: {Tuple2}")

if Tuple1 > Tuple2:
    print(f"{Tuple1} is greater than {Tuple2}")
elif Tuple1 < Tuple2:
    print(f"{Tuple1} is less than {Tuple2}")
else:
    print(f"{Tuple1} is equal to {Tuple2}")
```

14. Filter Tuples

Write a code to filter out students with scores less than 90 from a given a list of tuples.

Given:

students = [('Alice', 85), ('Bob', 92), ('Charlie', 78)]

Expected Output:

Students with scores 90 or above: [('Bob', 92), ('David', 95)]

```python
students = [('Alice', 85), ('Bob', 92), ('Charlie', 78), ('David', 95)]
students_with_90_or_above = []
for student in students:
    if student[1]>=90:
        students_with_90_or_above.append(student)
print(students_with_90_or_above)
```

15. Map Tuples

Given:

t = (1, 2, 3, 4)

Expected Output:

Original tuple: (1, 2, 3, 4)
