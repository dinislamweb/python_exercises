2. Write a program to create a new string made of the middle three characters of an input string.
    * str1 = "JhonDipPeta" Output : Dip
    * str2 = "JaSonAy" Output : Son

```python
def newString(x):
    start_of_middle = len(x)//2 - 1
    end_of_middle_three = len(x)//2 + 2
    return x[start_of_middle:end_of_middle_three]
str1 = "JhonDipPeta"
str2 = "JaSonAy"
print(newString(str1))
print(newString(str2))
```

3. Given two strings, s1 and s2. Write a program to create a new string s3 by appending s2 in the middle of s1.
    * s1 = "Ault"
    * s2 = "Kelly"
    * Expected Output: AuKellylt

```python
s1 = "Ault"
s2 = "Kelly"
middle_of_s1 = len(s1)//2
s3 = s1[:middle_of_s1]+s2+s1[middle_of_s1: ]
print(s3)
```

```python
def append_middle(s1,s2):
    middle_of_s1 = len(s1)//2
    s3 = s1[:middle_of_s1]+s2+s1[middle_of_s1: ]
    print(s3)
s1 = "Ault"
s2 = "Kelly"
append_middle(s1,s2)
```

4. Given two strings, s1 and s2, write a program to return a new string made of s1 and s2’s first, middle, and last characters.
    * s1 = "America"
    * s2 = "Japan"
    * Expected Output: AJrpan

```python
def newString(s1,s2):
    first_char = s1[0]+s2[0]
    middle_char = s1[len(s1)//2]+s2[len(s2)//2]
    last_char = s1[-1]+s2[-1]
    newString = first_char + middle_char + last_char
    return newString
s1 = "America"
s2 = "Japan"
print(newString)
```
