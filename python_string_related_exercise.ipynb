2. Write a program to create a new string made of the middle three characters of an input string.
    * str1 = "JhonDipPeta" Output : Dip
    * str2 = "JaSonAy" Output : Son

```python
def newString(x):
    start_of_middle = len(x)//2 - 1
    end_of_middle_three = len(x)//2 + 2
    return x[start_of_middle:end_of_middle_three]
str1 = "JhonDipPeta"
str2 = "JaSonAy"
print(newString(str1))
print(newString(str2))
```

4. Given two strings, s1 and s2, write a program to return a new string made of s1 and s2’s first, middle, and last characters.
    * s1 = "America"
    * s2 = "Japan"
    * Expected Output: AJrpan

```python
def MadeNewString(s1,s2):
    first_char = s1[0]+s2[0]
    middle_char = s1[len(s1)//2]+s2[len(s2)//2]
    last_char = s1[-1]+s2[-1]
    newString = first_char + middle_char + last_char
    return newString
s1 = "America"
s2 = "Japan"
print(MadeNewString(s1,s2))
```

5. Count all letters, digits, and special symbols from a given string
     * str1 = "P@#yn26at^&i5ve"
     * Expected Outcome:

     * Total counts of chars, digits, and symbols 

Chars = 8 
Digits = 3 
Symbol = 4

```python
def count_letter_digit_symbol(s):
    letters = 0
    digits = 0
    symbols = 0
    for i in s:
        if i.isalpha():
            letters = letters + 1
        elif i.isdigit():
            digits = digits + 1
        else:
            symbols = symbols + 1
    print(f"Chars = {letters} Digits = {digits} Symbol = {symbols}")
str1 = "P@#yn26at^&i5ve"
count_letter_digit_symbol(str1)
```

7. Write a program to check if two strings are balanced. For example, strings s1 and s2 are balanced if all the characters in the s1 are present in s2. The character’s position doesn’t matter.

Case 1:
- s1 = "Yn"
- s2 = "PYnative"
- Expected Output:
True

Case 2:
- s1 = "Ynf"
- s2 = "PYnative"
- Expected Output:
False

```python
def balancedString(s1,s2):
    for char in s1:
        if char in s2:
            continue
        else:
            return False
    return True
s1 = "Yna"
s2 = "PYnative"
print(balancedString(s1,s2))
```

9. Given a string s1, write a program to return the sum and average of the digits that appear in the string, ignoring all other characters.
    * str1 = "PYnative29@#8496"
- Expected Outcome:
- Sum is: 38 Average is  6.333333333333333

```python
def findAvgOfDigit(str1):
    sum1 = 0
    count = 0
    for digit in str1:
        if digit.isdigit():
            count = count + 1
            sum1 = sum1 + int(digit)
    avg = sum1 / count
    return avg
str1 = "PYnative29@#8496"
print(findAvgOfDigit(str1))
```

10. Write a program to count occurrences of all characters within a string
- str1 = "Apple"
- Expected Outcome:

{'A': 1, 'p': 2, 'l': 1, 'e': 1}

```python
str1 = "Apple"
char_dict = dict()

for char in str1:
    count = str1.count(char)
    # add / update the count of a character
    char_dict[char] = count
print('Result:', char_dict)   
```

```python
from collections import Counter
str1 = "Apple"
char_dict = Counter(str1)

print("Result:", dict(char_dict))
```
